```python
import torch
from PIL import Image
import torchvision.transforms as transforms
import numpy as np

```

```python
blend = Image.open('/mnt/d/Research/YTMT-Strategy/real20_420/blended/39.jpg').convert('RGB')
trans = Image.open('/mnt/d/Research/YTMT-Strategy/real20_420/transmission_layer/39.jpg').convert('RGB')

blend.show()
trans.show()
```

```python
transform = transforms.ToTensor()

blend = transform(blend)
trans = transform(trans)

```

```python
print(blend)
```

```python
R = blend-trans
```

```python
print(R)
```

```python
transform = transforms.ToPILImage()
R = transform(R)
R.show()
```

```python
R = np.asarray(R)/255
R = np.clip(R,0,1)
print(R.shape)
R = R * 255.0
print(R.shape)
R = Image.fromarray(R.astype(np.uint8))
R.show()
```
